### 1. **Simple Linear Regression vs. Multiple Linear Regression**

**Simple Linear Regression:**
- Simple Linear Regression models the relationship between one independent variable (predictor) and one dependent variable (outcome). The model assumes a linear relationship between the two variables.

  **Equation:**
  ```
  outcome = intercept + coefficient_X * predictor_X
  ```
  where:
  - `intercept` is the expected value of the outcome when `predictor_X = 0`,
  - `coefficient_X` is the coefficient for the predictor `predictor_X`,
  - `predictor_X` is the independent variable.

**Multiple Linear Regression:**
- Multiple Linear Regression extends Simple Linear Regression by including two or more independent variables to predict the dependent variable. This allows modeling more complex relationships between the outcome and multiple predictors.

  **Equation:**
  ```
  outcome = intercept + coefficient_X * predictor_X + coefficient_Y * predictor_Y
  ```
  where:
  - `intercept` is the expected value of the outcome when `predictor_X = 0` and `predictor_Y = 0`,
  - `coefficient_X` and `coefficient_Y` are the coefficients for the predictors `predictor_X` and `predictor_Y`,
  - `predictor_X` and `predictor_Y` are the independent variables.

**Benefit of Multiple Linear Regression Over Simple Linear Regression:**
- Multiple Linear Regression can model more complex relationships by accounting for multiple predictors simultaneously, capturing interactions and the individual contribution of each variable. It is useful when multiple factors influence the outcome.

---

### 2. **Continuous Variable vs. Indicator Variable in Simple Linear Regression**

**Continuous Variable in Simple Linear Regression:**
- A continuous variable is a predictor that can take any value within a range. The relationship between the predictor and outcome is modeled as a straight line.

  **Equation:**
  ```
  outcome = intercept + coefficient_X * predictor_X
  ```
  where `predictor_X` is continuous.

**Indicator Variable in Simple Linear Regression:**
- An indicator variable is a categorical variable that is converted into a binary form (0 or 1). It is used to capture the effect of different categories on the outcome.

  **Equation:**
  ```
  outcome = intercept + coefficient_1 * indicator(Category_A)
  ```
  where `indicator(Category_A)` is the indicator variable representing the presence of Category A.

**Difference:**
- When using a continuous variable, the model predicts a change in the outcome for every unit increase in the predictor. When using an indicator variable, the model compares the outcome between different categories.

---

### 3. **Introducing an Indicator Variable with a Continuous Variable in Multiple Linear Regression**

When both a continuous variable and an indicator variable are included in the model, the outcome is influenced by both predictors. The continuous variable adds a linear relationship, and the indicator variable introduces a categorical shift (0 or 1 effect).

**Equation:**
```
outcome = intercept + coefficient_X * predictor_X + coefficient_1 * indicator(Category_A)
```
- **Interpretation:** The `intercept` is the baseline outcome for Category A, `coefficient_X` represents the effect of the continuous variable `predictor_X`, and `coefficient_1` captures the difference between Category A and the baseline.

**Expected Behavior:**
- For a continuous variable `predictor_X`, the model estimates the change in the outcome for each unit change in `predictor_X`.
- For the indicator variable, it shifts the outcome based on the category (e.g., being in Category A versus the baseline category).

---

### 4. **Effect of Adding an Interaction Between a Continuous and an Indicator Variable in Multiple Linear Regression**

When an interaction term between a continuous variable and an indicator variable is included in the model, it allows the effect of the continuous variable to vary depending on the category represented by the indicator variable.

**Equation:**
```
outcome = intercept + coefficient_X * predictor_X + coefficient_1 * indicator(Category_A) + coefficient_2 * (predictor_X * indicator(Category_A))
```
- **Interpretation:** The `coefficient_2` represents how the effect of `predictor_X` differs between Category A and the baseline category.

**Expected Behavior:**
- The relationship between `predictor_X` and the outcome is no longer constant. Instead, it changes depending on the category of the indicator variable, allowing for more nuanced modeling.

---

### 5. **Multiple Linear Regression with Only Indicator Variables from a Non-Binary Categorical Variable**

When a non-binary categorical variable is used, it is typically converted into multiple indicator variables (dummy variables). Each indicator variable represents a category, with the reference category omitted.

**Equation:**
```
outcome = intercept + coefficient_1 * indicator(Category_1) + coefficient_2 * indicator(Category_2) + ... + coefficient_(n-1) * indicator(Category_n-1)
```
- Each `indicator(Category_i)` represents a specific category, and the `intercept` represents the outcome for the baseline category.

**Interpretation:**
- The `intercept` represents the outcome for the baseline category, and each `coefficient_i` represents the difference in the outcome between Category `i` and the baseline category.

**Role of the "Baseline" Group:**
- The baseline group serves as the reference point, and its outcome is captured by the intercept. The coefficients for the indicator variables show how the outcome for other categories compares to the baseline.

**Binary Variable Encodings:**
- Indicator variables are binary (0 or 1), and only `n - 1` categories need to be encoded to represent an `n`-category variable (the omitted category serves as the baseline).

**Example Use-Case:**
- If the categorical variable is "Region" with three categories (North, South, West), the model would use two indicator variables (e.g., `indicator(North)`, `indicator(South)`) with the West region as the baseline. The coefficients `coefficient_1` and `coefficient_2` represent how the outcome differs for North and South regions compared to the baseline West region.

Outcome and Predictor Variables:
Outcome Variable: In this scenario, the outcome variable is the effectiveness of the advertising campaign. This could be measured by the resulting sales, engagement, or any other metric that reflects how successful the ad campaigns are at driving business results.
Predictor Variables:
The TV advertising budget (a continuous variable representing the amount of money spent on TV ads).
The online advertising budget (also a continuous variable representing the amount of money spent on online ads).
Given that the effectiveness of the TV ad might depend on the amount spent on online advertising and vice versa, there could be an interaction effect between the two advertising budgets that needs to be accounted for in the model. An interaction means that the effect of one variable (e.g., TV budget) on the outcome may depend on the value of the other variable (e.g., online budget).

Models with and without Interaction:
1. Without Interaction (Additive Model)

In this model, we assume that the effects of the TV ad and online ad budgets on the outcome are independent of each other. That is, increasing the TV budget has a fixed effect on the outcome, regardless of the amount spent on online ads, and vice versa.

Linear form (without interaction):

Outcome = intercept + coefficient_TV * TV_ad_budget + coefficient_Online * Online_ad_budget
Interpretation:
intercept: Represents the baseline effectiveness of the advertising campaign when both TV and online ad budgets are zero.
coefficient_TV: The effect on the outcome of increasing the TV ad budget by one unit, assuming the online ad budget stays the same.
coefficient_Online: The effect on the outcome of increasing the online ad budget by one unit, assuming the TV ad budget stays the same.
This model assumes no interaction between the TV and online ad budgets, meaning that the combined effect of both budgets is simply the sum of their individual effects.

2. With Interaction (Synergistic Model)

In this model, the effect of one predictor depends on the value of the other predictor. For example, the impact of the TV ad budget on the outcome may change depending on how much is spent on online ads.

Linear form (with interaction):

Outcome = intercept + coefficient_TV * TV_ad_budget + coefficient_Online * Online_ad_budget + coefficient_Interaction * (TV_ad_budget * Online_ad_budget)
Interpretation:
intercept: Represents the baseline effectiveness when both budgets are zero.
coefficient_TV: The effect of TV ad budget on the outcome, holding the online ad budget constant.
coefficient_Online: The effect of online ad budget on the outcome, holding the TV ad budget constant.
coefficient_Interaction: The effect of the interaction between TV and online ad budgets, capturing how the relationship between one ad budget and the outcome changes based on the other ad budget.
This model considers that the effect of increasing one advertising budget (TV or online) depends on the level of the other budget. The interaction term allows for a more flexible model, capturing cases where the combined effect of TV and online ads is more than just the sum of their individual effects.

Making Predictions from the Models:
Without Interaction:
The prediction for the outcome is simply based on the individual effects of TV and online ad budgets. The effect of increasing either budget is fixed and does not depend on the level of the other budget. So, the prediction is a linear combination of the two predictors.
With Interaction:
The prediction takes into account both the individual effects of TV and online ad budgets, as well as how the effect of one predictor changes depending on the level of the other predictor. The interaction term adjusts the prediction to reflect this interdependence between the predictors.
Example:

If you have a high TV budget and a low online budget, the model without interaction will simply add the effects of each budget on the outcome.
However, with interaction, the model will consider how the high TV budget might interact with the low online budget, and vice versa, adjusting the prediction accordingly.
Binary Variables for Ad Budgets:
Now, instead of continuous variables, suppose the advertising budgets are categorized as either "high" or "low" (binary variables). In this case, the model needs to account for the categorical nature of the predictors.

3. Without Interaction (Additive Model with Binary Variables)

When the budgets are binary, the model will predict the outcome based on whether each budget is high or low. There is no interaction between the two budgets in this model.

Linear form (without interaction, binary variables):

Outcome = intercept + coefficient_TV * 1(TV_ad_budget = High) + coefficient_Online * 1(Online_ad_budget = High)
1(TV_ad_budget = High) is an indicator variable that equals 1 if the TV budget is high, and 0 if it is low.
1(Online_ad_budget = High) is an indicator variable that equals 1 if the online budget is high, and 0 if it is low.
The coefficients represent the change in the outcome when the corresponding advertising budget is high (relative to low).

4. With Interaction (Synergistic Model with Binary Variables)

In this model, we consider how the combined effect of high TV and high online budgets might differ from the individual effects of each.

Linear form (with interaction, binary variables):

Outcome = intercept + coefficient_TV * 1(TV_ad_budget = High) + coefficient_Online * 1(Online_ad_budget = High) + coefficient_Interaction * (1(TV_ad_budget = High) * 1(Online_ad_budget = High))
coefficient_Interaction represents the additional effect when both TV and online ad budgets are high at the same time. This allows for the possibility that the effect of high TV ads might depend on whether the online budget is also high.

1. Logistic Regression Setup:
You mentioned that we need to predict a binary outcome variable. In logistic regression, the outcome is binary (e.g., "yes" or "no"), so let's say you have a binary variable like "social_activity_participation" (1 for participation and 0 for no participation) in your dataset.

We'll combine continuous and binary predictors in the model. Here's how you would fit the logistic regression model using smf.logit from statsmodels.formula.api:

import pandas as pd
import statsmodels.formula.api as smf

### Replace with the correct dataset URL or data loading method
url = "your_dataset_url_here.csv"
data = pd.read_csv(url).fillna('None')

### Example transformation for a categorical variable to binary
### Let's assume 'income_level' is a categorical variable with 'high' and 'low' categories
data['high_income'] = (data['income_level'] == 'high').astype(int)

### Define the logistic regression model formula
### Let's say 'social_activity_participation' is the binary outcome variable
### We'll use continuous predictor 'age' and the binary predictor 'high_income'
log_reg_formula = 'social_activity_participation ~ age + high_income'

### Fit the logistic regression model
log_reg_fit = smf.logit(log_reg_formula, data=data).fit()

### Show the model summary
print(log_reg_fit.summary())
Explanation:
social_activity_participation is the binary outcome.
age is a continuous predictor.
high_income is a binary predictor based on whether the person's income level is "high" (1) or "low" (0).
2. Interpretation of Logistic Regression Results:
The output from log_reg_fit.summary() will show you the coefficients, which represent the log odds of the outcome. For logistic regression, interpreting the coefficients directly as you would in a linear regression is not as straightforward, but we can make it simpler for this exercise.

Intercept: The log odds of participating in social activities when all predictors are at their reference levels (e.g., for someone with age = 0 and high_income = 0).
Age coefficient: The change in log odds of participation in social activities for each additional year of age, holding income constant.
High_income coefficient: The change in log odds of participation for someone with high income (compared to low income), holding age constant.
Even though these are log odds, for simplicity, you can interpret the results as if you were interpreting linear regression coefficients. The key difference is that these coefficients represent changes in log odds, not direct changes in probability.

3. Making Predictions:
To make predictions from the model, you can use the predict() function. The prediction will give you the probability of the outcome (in this case, the probability of participating in social activities).

### Get the predicted probabilities
predicted_probabilities = log_reg_fit.predict(data)

### To classify as 1 or 0, you can set a threshold (commonly 0.5)
predicted_classes = (predicted_probabilities >= 0.5).astype(int)
Explanation:
predicted_probabilities is the probability of participating in social activities, which ranges between 0 and 1.
predicted_classes classifies people as either 1 (participating) or 0 (not participating), depending on whether their predicted probability is greater than or equal to 0.5.
4. Visualization with Plotly:
To visualize the results, you can use Plotly. We'll plot the data with the fitted "best fit lines" for both the additive and synergistic models.

Additive Model (no interaction):

For the additive model, we'll just consider the predictors independently.

import plotly.express as px

### Generate predictions from the additive model
data['additive_predictions'] = log_reg_fit.predict(data)

### Plot the data and the fitted line
fig = px.scatter(data, x='age', y='social_activity_participation', color='high_income',
                 labels={'age': 'Age', 'social_activity_participation': 'Social Activity Participation'},
                 title='Additive Model: Social Activity vs Age and Income')

### Add the fitted logistic curve to the plot
fig.add_scatter(x=data['age'], y=data['additive_predictions'], mode='lines', name='Fitted Line')

fig.show()
Explanation: This scatter plot shows how the social activity participation depends on age and income, with a fitted line based on the additive model. Here, high_income is treated as a categorical variable (colored differently).
Synergistic Model (with interaction):

For the synergistic model, we'll add an interaction term between age and high_income to the formula.

### Add interaction term between age and high_income
log_reg_interaction_formula = 'social_activity_participation ~ age * high_income'

### Fit the logistic regression model with the interaction
log_reg_interaction_fit = smf.logit(log_reg_interaction_formula, data=data).fit()

### Generate predictions from the synergistic model
data['interaction_predictions'] = log_reg_interaction_fit.predict(data)

### Plot the data and the fitted line for the synergistic model
fig = px.scatter(data, x='age', y='social_activity_participation', color='high_income',
                 labels={'age': 'Age', 'social_activity_participation': 'Social Activity Participation'},
                 title='Synergistic Model: Social Activity vs Age and Income')

### Add the fitted logistic curve to the plot
fig.add_scatter(x=data['age'], y=data['interaction_predictions'], mode='lines', name='Fitted Line')

fig.show()
Explanation: This plot shows how the social activity participation depends on age and income, with a fitted line based on the synergistic model, considering the interaction term. The interaction term allows the effect of age on participation to differ based on income.
5. Interpretation of Visualization:
Additive Model: The fitted line in the plot should show how age affects the likelihood of social activity participation, assuming the effect of income is constant.
Synergistic Model: The interaction allows for a different slope for the "high_income" and "low_income" groups. You can interpret the curves as showing that the effect of age on participation is different depending on whether someone has a high or low income.

Interpreting R-squared = 0.176
An R-squared of 0.176 means that the model explains 17.6% of the variability in the outcome variable (e.g., HP in this case).

Interpreting Coefficients > 10
Coefficients greater than 10 suggest that some predictors have a strong relationship with the outcome y

Why Coefficients Can Be Large While R-squared Is Low:

- Strong local effects: Even if individual predictors have large coefficients, they might explain only a small portion of the overall variability in y.
- Few significant predictors: If only a small subset of predictors significantly affect y, their large coefficients won’t contribute enough to increase R-squared substantially, especially in noisy data.
- Unexplained variability: A large amount of variation in y might be unrelated to the predictors in the model, driving down R-squared.

The low r-squared means that the model struggles to explain most of the outcome's variability.
The large coefficients with small p-values mean that the predictors with large coefficients have a statistically significant relationship with the outcome. 

This means that while individual predictors may strongly influence the outcome locally, their combined explanatory power is limited relative to the total variability in the data. Thus, the model can have statistically significant predictors without providing a strong overall explanation of the outcome.

The apparent contradiction lies in the difference between what R-squared and p-values represent. R-squared measures the proportion of the total variability in the outcome that the model explains, and a low R-squared (17.6%) suggests the model leaves much of the variation unexplained. On the other hand, large coefficients with strong evidence (low p-values) indicate that specific predictors significantly influence the outcome while accounting for other predictors. These findings are not contradictory because R-squared reflects the model’s overall explanatory power, while p-values test whether each predictor has a statistically significant association with the outcome. A model can include predictors that are statistically significant (indicating strong local effects) but still fail to explain most of the variability in the data, often due to omitted variables or substantial random noise.

````python
import numpy as np
import pandas as pd
import statsmodels.formula.api as smf
import plotly.express as px
import plotly.graph_objects as go
from sklearn.model_selection import train_test_split  # Import train_test_split

# Ensure your dataset is loaded
# Replace this with the actual loading code for your 'pokeaman' dataset
# For example: pokeaman = pd.read_csv('pokeaman.csv')
pokeaman = pd.DataFrame({
    "HP": np.random.randint(50, 150, 100),
    "Attack": np.random.randint(50, 150, 100),
    "Defense": np.random.randint(50, 150, 100),
})

# Placeholder for the number of repetitions
reps = 100

# Arrays to store R-squared values
in_sample_Rsquared = np.zeros(reps)
out_of_sample_Rsquared = np.zeros(reps)

# Model linear form (for Model 3)
linear_form = 'HP ~ Attack + Defense'

# Run repeated train-test splits and model fitting
for i in range(reps):
    # 50-50 random train-test split
    pokeaman_train, pokeaman_test = train_test_split(
        pokeaman, train_size=0.5, random_state=None
    )
    
    # Fit the model on the training data
    model_fit = smf.ols(formula=linear_form, data=pokeaman_train).fit()
    
    # Record "in-sample" R-squared
    in_sample_Rsquared[i] = model_fit.rsquared
    
    # Record "out-of-sample" R-squared
    yhat_test = model_fit.predict(pokeaman_test)
    y_test = pokeaman_test.HP
    out_of_sample_Rsquared[i] = np.corrcoef(y_test, yhat_test)[0, 1] ** 2

# Create a DataFrame for visualization
df = pd.DataFrame({
    "In Sample Performance (R-squared)": in_sample_Rsquared,
    "Out of Sample Performance (R-squared)": out_of_sample_Rsquared
})

# Create scatter plot
fig = px.scatter(
    df,
    x="In Sample Performance (R-squared)",
    y="Out of Sample Performance (R-squared)",
    title="In-Sample vs Out-of-Sample Performance",
    labels={"x": "In-Sample R-squared", "y": "Out-of-Sample R-squared"}
)
fig.add_trace(go.Scatter(x=[0, 1], y=[0, 1], mode="lines", name="y=x"))
fig.show(renderer='png')

````
