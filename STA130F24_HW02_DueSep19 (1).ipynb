### 3. Streamlined Code <br>

```python
import numpy as np

reps = 100000  # Number of repetitions
wins = 0  # Track the number of wins

for _ in range(reps):
    winning_door = np.random.choice([1, 2, 3])  # Randomly assign the winning door
    my_choice = np.random.choice([1, 2, 3])  # Randomly make an initial choice

    # Monty opens a door revealing a goat
    remaining_doors = [door for door in [1, 2, 3] if door != my_choice and door != winning_door]
    revealed_door = np.random.choice(remaining_doors)

    # The contestant switches to the remaining unopened door
    switch_choice = [door for door in [1, 2, 3] if door != my_choice and door != revealed_door][0]

    # Count wins if the switched choice is the winning door
    if switch_choice == winning_door:
        wins += 1

# Calculate the probability of winning by switching
probability_of_winning = wins / reps
print(f"Probability of winning by switching: {probability_of_winning:.4f}")
```
